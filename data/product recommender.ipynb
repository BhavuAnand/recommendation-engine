```python
product = pd.read_csv('myproducts.csv', encoding= 'unicode_escape', header=0)
```

```python
product['tags'] = product['subCategory'] + product['articleType'] + product['productDisplayName'] + product['keywords']
```

```python
new_df = product[['id','productDisplayName','homepage','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower() if type(x) is str else 'empty') 
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(products):
    products_index = new_df[new_df['productDisplayName'] == products].index[0]
    distances = similarity[products_index]
    products_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in products_list:
        print(new_df.iloc[i[0]].productDisplayName)
    for i in products_list:
        print(new_df.iloc[i[0]].homepage)
```

```python
recommend('Lotus Herbals Tea Tree Face Wash')
```
